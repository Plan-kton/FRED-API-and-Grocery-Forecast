# üìä Matplotlib Cheat Sheet

Matplotlib is a powerful library for creating static, animated, and interactive visualizations in Python.

## üîπ Importing Matplotlib
```python
import matplotlib.pyplot as plt
import numpy as np
```

---
## **1Ô∏è‚É£ Simple (State-Based) Approach**
The **state-based interface** uses `plt.` directly and lets Matplotlib handle figure and axes creation automatically. This is best for **quick, simple plots**.

### üìà **Basic Line Plot**
```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='sin(x)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Line Plot')
plt.legend()
plt.show()
```

### üìä **Bar Chart**
```python
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

plt.bar(categories, values, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.show()
```

### üü¢ **Scatter Plot**
```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='green', alpha=0.6)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')
plt.show()
```

### üìä **Histogram**
```python
data = np.random.randn(1000)

plt.hist(data, bins=30, color='purple', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.show()
```

---
## **2Ô∏è‚É£ Object-Oriented (OOP) Approach**
The **OOP interface** gives more control by explicitly managing **Figure** and **Axes** objects. This is best for **multiple subplots and advanced customization**.

### **üìà Line Plot (OOP Approach)**
```python
fig, ax = plt.subplots(figsize=(8,5))  # Create figure and axis

x = np.linspace(0, 10, 100)
y = np.sin(x)

ax.plot(x, y, label='sin(x)')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Basic Line Plot')
ax.legend()

plt.show()
```

### **üìä Bar Chart (OOP Approach)**
```python
fig, ax = plt.subplots()

categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

ax.bar(categories, values, color='skyblue')
ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Bar Chart Example')

plt.show()
```

### **üìä Multiple Subplots**
```python
fig, ax = plt.subplots(2, 2, figsize=(8, 6))

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

ax[0, 0].plot(x, y1, 'r')
ax[0, 0].set_title('Sine Wave')

ax[0, 1].plot(x, y2, 'b')
ax[0, 1].set_title('Cosine Wave')

ax[1, 0].hist(np.random.randn(100), bins=20, color='g')
ax[1, 0].set_title('Histogram')

ax[1, 1].scatter(np.random.rand(50), np.random.rand(50), color='k')
ax[1, 1].set_title('Scatter Plot')

plt.tight_layout()
plt.show()
```

### üé® **Customizing Plots (OOP Approach)**
```python
fig, ax = plt.subplots()

x = np.linspace(0, 10, 100)
y = np.sin(x)

ax.plot(x, y, linestyle='--', linewidth=2, color='red', marker='o', markersize=5)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Customized Plot')

plt.show()
```

### üîÑ **Saving a Figure**
```python
fig, ax = plt.subplots()
ax.plot(x, y)
fig.savefig('plot.png', dpi=300, bbox_inches='tight')
```

### üèÅ **Show All Open Figures**
```python
plt.show()
```

---
## **üöÄ Key Takeaways**
| Feature | Simple (`plt.`) | Object-Oriented (`fig, ax`) |
|---------|----------------|--------------------------|
| **Best for** | Quick, simple plots | Advanced, multiple subplots |
| **Flexibility** | Limited customization | More control over figures & axes |
| **Multiple subplots** | ‚ùå No | ‚úÖ Yes |
| **Explicit control** | ‚ùå No | ‚úÖ Yes |
| **Ease of use** | ‚úÖ Yes | ‚ùå More code |

Choose **Simple (`plt.`)** for **quick plots**, and **Object-Oriented (`fig, ax`)** for **better control and multiple subplots**! üöÄ


# üìä Seaborn Cheat Sheet

Seaborn is a Python library for statistical data visualization, built on Matplotlib. It provides **beautiful, high-level functions** for plotting different types of data.

---
## **1Ô∏è‚É£ Importing Seaborn**
```python
import seaborn as sns
import matplotlib.pyplot as plt
```

---
## **2Ô∏è‚É£ Loading Example Datasets**
```python
# Load built-in dataset
df = sns.load_dataset("tips")  # Example dataset
df.head()  # View first few rows
```

---
## **3Ô∏è‚É£ Basic Plot Customization**
```python
sns.set_style("darkgrid")  # Set background style
sns.set_palette("pastel")  # Set color palette
```

---
## **4Ô∏è‚É£ Distribution Plots**
### **üîπ Histogram (`histplot`)**
```python
sns.histplot(df["total_bill"], bins=20, kde=True)
plt.show()
```
### **üîπ Kernel Density Estimation (`kdeplot`)**
```python
sns.kdeplot(df["total_bill"], shade=True)
plt.show()
```

---
## **5Ô∏è‚É£ Categorical Plots**
### **üîπ Bar Plot (`barplot`)**
```python
sns.barplot(x="day", y="total_bill", data=df)
plt.show()
```
### **üîπ Box Plot (`boxplot`)**
```python
sns.boxplot(x="day", y="total_bill", data=df)
plt.show()
```
### **üîπ Violin Plot (`violinplot`)**
```python
sns.violinplot(x="day", y="total_bill", data=df)
plt.show()
```

---
## **6Ô∏è‚É£ Relational Plots**
### **üîπ Scatter Plot (`scatterplot`)**
```python
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=df)
plt.show()
```
### **üîπ Line Plot (`lineplot`)**
```python
sns.lineplot(x="size", y="total_bill", data=df)
plt.show()
```

---
## **7Ô∏è‚É£ Pairwise Plots**
### **üîπ Pair Plot (`pairplot`)**
```python
sns.pairplot(df, hue="sex")
plt.show()
```
### **üîπ Heatmap (`heatmap`)**
```python
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.show()
```

---
## **8Ô∏è‚É£ Regression Plots**
### **üîπ Regression Plot (`regplot`)**
```python
sns.regplot(x="total_bill", y="tip", data=df)
plt.show()
```
### **üîπ LM Plot (`lmplot`)**
```python
sns.lmplot(x="total_bill", y="tip", hue="sex", data=df)
plt.show()
```

---
## **9Ô∏è‚É£ Styling & Customization**
```python
sns.set_theme(style="whitegrid", palette="muted")  # Set theme
plt.figure(figsize=(10, 6))  # Change figure size
sns.despine()  # Remove top and right borders
```

---
## **üîü Saving Plots**
```python
sns.histplot(df["total_bill"], bins=20)
plt.savefig("histogram.png", dpi=300)
```

---

| Function | Use |

````python
fig, ax = plt.subplots(2, 2, figsize=(8, 6))  #the ax creates a numpy array that is used to identify each of the axis and charts, eg, ax[0 ,0] is first row and first chart

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

ax[0, 0].plot(x, y1, 'r')
ax[0, 0].set_title('Sine Wave')

ax[0, 1].plot(x, y2, 'b')
ax[0, 1].set_title('Cosine Wave')

ax[1, 0].hist(np.random.randn(100), bins=20, color='g')
ax[1, 0].set_title('Histogram')

ax[1, 1].scatter(np.random.rand(50), np.random.rand(50), color='k')
ax[1, 1].set_title('Scatter Plot')

plt.tight_layout()
plt.show()
````
