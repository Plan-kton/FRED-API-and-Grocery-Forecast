## ðŸ“Œ 1. Headers
Use `#` for headings. More `#` means a smaller heading.

```markdown
# H1 - Largest Header
## H2 - Second Largest
### H3 - Medium Header
#### H4 - Smaller Header
##### H5 - Tiny Header
###### H6 - Smallest Header
```

# H1 - Largest Header  
## H2 - Second Largest  
### H3 - Medium Header  
#### H4 - Smaller Header  
##### H5 - Tiny Header  
###### H6 - Smallest Header
