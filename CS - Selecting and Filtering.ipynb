# 📝 Python Selection & Filtering Cheat Sheet

Selecting and filtering data efficiently is essential when working with different data structures in Python. This cheat sheet categorizes selection and filtering methods by **lists**, **dictionaries**, **sets**, **tuples**, **pandas Series**, and **pandas DataFrames**.

---
## 🔹 Selecting & Filtering in Lists
```python
# Example list
numbers = [10, 15, 20, 25, 30]

# ✅ Select by index
first_element = numbers[0]  # 10
last_element = numbers[-1]  # 30

# ✅ Select a slice
subset = numbers[1:4]  # [15, 20, 25]

# ✅ Select using list comprehension
filtered = [num for num in numbers if num > 20]
```

---
## 🔹 Selecting & Filtering in Dictionaries
```python
# Example dictionary
data = {'a': 10, 'b': 25, 'c': 30, 'd': 15}

# ✅ Select by key
value_b = data['b']  # 25

# ✅ Get all keys or values
keys = data.keys()   # dict_keys(['a', 'b', 'c', 'd'])
values = data.values()  # dict_values([10, 25, 30, 15])

# ✅ Filter dictionary by values
filtered = {k: v for k, v in data.items() if v > 20}
```

---
## 🔹 Selecting & Filtering in Sets
```python
# Example set
numbers = {10, 15, 20, 25, 30}

# ✅ Check membership
exists = 20 in numbers  # True

# ✅ Filter using set comprehension
filtered = {num for num in numbers if num > 20}
```

---
## 🔹 Selecting & Filtering in Tuples
```python
# Example tuple
numbers = (10, 15, 20, 25, 30)

# ✅ Select by index
first_element = numbers[0]  # 10
last_element = numbers[-1]  # 30

# ✅ Select a slice
subset = numbers[1:4]  # (15, 20, 25)
```

---
## 🔹 Selecting & Filtering in Pandas Series
```python
import pandas as pd
import numpy as np

# Example Series
data = pd.Series([10, 15, np.nan, 25, 30], index=['a', 'b', 'c', 'd', 'e'])

# ✅ Select by label
d = data['d']  # 25

# ✅ Select by position
d = data.iloc[3]  # 25

# ✅ Filter values greater than 20
data_filtered = data[data > 20]

# ✅ Drop NaN values
data_filtered = data.dropna()
```

---
## 🔹 Selecting & Filtering in Pandas DataFrames
```python
# Example DataFrame
df = pd.DataFrame({
    'A': [10, 15, np.nan, 25, 30],
    'B': ['x', 'y', 'z', 'w', 'v']
})

# ✅ Select a single column
df_a = df['A']

# ✅ Select multiple columns
df_subset = df[['A', 'B']]

# ✅ Select by row index
df_first_row = df.iloc[0]
df_last_row = df.iloc[-1]

# ✅ Select by label
df_label_row = df.loc[0]

# ✅ Filter rows where column 'A' is greater than 20
df_filtered = df[df['A'] > 20]

# ✅ Select rows where column1 > 10
df.loc[df['column1'] > 10]

# ✅ Filter using multiple conditions
df_filtered = df[(df['A'] > 10) & (df['B'] == 'y')]

# ✅ Select rows where ANY column is NaN
df_filtered = df[df.isna().any(axis=1)]

# ✅ Select rows where ALL columns are NaN
df_filtered = df[df.isna().all(axis=1)]
```

---
## 🔹 Filtering with Query (Pandas)
```python
# ✅ Using query() to filter
df_filtered = df.query("A > 20")
```

---
## 🔹 Selecting & Filtering Index in DataFrames
```python
# ✅ Filter using index
df_filtered = df[df.index > 2]
```

This cheat sheet provides a **quick reference** for selecting and filtering across different Python data structures. 🚀 Happy coding! 😊
